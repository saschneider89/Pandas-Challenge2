```python
# Dependencies
import pandas as pd
```

```python
#csv_path = "Resources/purchase_data.csv"
#Import the purchase_data.csv file as DataFrame
purchase_data = pd.read_csv("Resources/purchase_data.csv",encoding="utf-8")
purchase_data.head()
```

```python
purchase_data.dtypes
```

```python
purchase_data = purchase_data.rename(columns={"SN":"Players"})
```

```python
#Player Count
# Calculate the number of unique Players in the DataFrame
Total_Players = purchase_data["Players"].nunique()
Total_Players_df = pd.DataFrame({"Total Players":[Total_Players]})
Total_Players_df.head()
```

```python
#Purchasing Analysis
#Unique_Items = len(purchase_data["Item Name"].unique())
#Total_Revenue = sum(purchase_data["Price"])
Unique_Items = purchase_data["Item ID"].nunique()
Average_Price = round(purchase_data["Price"].mean(),2)
Number_Purchases = purchase_data["Item Name"].count()
Total_Revenue = purchase_data["Price"].sum()
Purchasing_Analysis = pd.DataFrame({"Number of Unique Items":[Unique_Items],
                                    "Average Price":[Average_Price],
                                    "Number of Purchases":[Number_Purchases],
                                    "Total Revenue":[Total_Revenue]})
#Cleaner format
Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${:,.2f}".format)
Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Total Revenue"].map("${:,.2f}".format)
Purchasing_Analysis
```

```python
#Gender Demographics
Gender_Groups = purchase_data.groupby(["Gender"])
Gender_Count = Gender_Groups_df["Players"].nunique()
Gender_Percent = Gender_Count / Total_Players * 100
Gender_Demographics = pd.DataFrame({"Total Count": Gender_Count,
                         "Percentage of Players": Gender_Percent})
#Cleaner format
Gender_Demographics["Percentage of Players"] = Gender_Demographics["Percentage of Players"].map("{:,.2f}%".format)
Gender_Demographics.sort_values(["Total Count"], ascending=False)
```

```python
#Purchasing Analysis (Gender)
Purchase_Count = Gender_Groups["Purchase ID"].count()
Average_Purchase_Price = Gender_Groups["Price"].mean()
Total_Purchase_Value = Gender_Groups["Price"].sum()
Average_Purchase_Person = Total_Purchase_Value / Gender_Count
Purchasing_Analysis = pd.DataFrame({"Purchase Count": Purchase_Count,
                                   "Average Purchase Price": Average_Purchase_Price,
                                   "Total Purchase Value": Total_Purchase_Value,
                                   "Avg Total Purchase Purchase per Person": Average_Purchase_Person})
#Cleaner format
Purchasing_Analysis["Average Purchase Price"] = Purchasing_Analysis["Average Purchase Price"].map("${:,.2f}".format)
Purchasing_Analysis["Avg Total Purchase Purchase per Person"] = Purchasing_Analysis["Avg Total Purchase Purchase per Person"].map("${:,.2f}".format)
Purchasing_Analysis.head()
```

```python
#Age Demographics, (See ex: 4.3.4 Binning Ted)
#Create bins/labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#Slice data and place into bins
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels= group_labels)
# Place the data series into a new column inside of the DataFrame
Age_Group_df = purchase_data.groupby(["Age Groups"])
```

```python
#Simple calculations
Age_Count = Age_Group_df["Players"].nunique()
Age_Percentage = Age_Count / Total_Players * 100
#Summary dataframe
Age_Demographic_df = pd.DataFrame({"Total Count": Age_Count,
                                  "Percentage of Players": Age_Percentage})
#Cleaner format
Age_Demographic_df["Percentage of Players"] = Age_Demographic_df["Percentage of Players"].map("{:.2f}%".format)
print(Age_Demographic_df)
```

```python
#Purchasing Analysis (Age)
#Simple Calculations
Purchase_Count = Age_Group_df["Purchase ID"].count()
Average_Purchase_Price = Age_Group_df["Price"].mean()
Total_Purchase_Value = Age_Group_df["Price"].sum()
Average_Purchase_Person = Total_Purchase_Value / Age_Count
#Summary dataframe
Purchasing_Age_df = pd.DataFrame({"Purchase Count": Purchase_Count,
                                 "Average Purchase Price": Average_Purchase_Price,
                                 "Total Purchase Value": Total_Purchase_Value,
                                 "Avg Total Purchase per Person": Average_Purchase_Person})
#Cleaner format
Purchasing_Age_df["Average Purchase Price"] = Purchasing_Age_df["Average Purchase Price"].map("${:,.2f}".format)
Purchasing_Age_df["Total Purchase Value"] = Purchasing_Age_df["Total Purchase Value"].map("${:,.2f}".format)
Purchasing_Age_df["Avg Total Purchase per Person"] = Purchasing_Age_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
Purchasing_Age_df
```
